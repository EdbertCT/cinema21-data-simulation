```
import pandas as pd
```

```
path = "https://raw.githubusercontent.com/ridhof/cinema21-data-simulation/master/data%20survey%20raw/keluar.csv"
output_path = "data_keluar.csv"
raw_data = pd.read_csv(path)
raw_data
```

```
raw_data.loc[0].values.tolist()[0].split(";")
```

```
lap = []
lap_time = []
total_time = []

for row in raw_data.values.tolist():
  temp = row[0].split(";")
  lap.append(temp[0])
  lap_time.append(temp[1])
  total_time.append(temp[2])

data = pd.DataFrame(data={'Lap': lap, 'Lap_time': lap_time, 'Total_time': total_time})
data
```

```
masuk_antrian = []
mulai_layanan = []
selesai_layanan = []

for i in range(0, len(data), 3):

  data_masuk_antrian = data.loc[i].values.tolist()[2]
  data_mulai_layanan = data.loc[i + 1].values.tolist()[2]
  data_selesai_layanan = data.loc[i + 2].values.tolist()[2]

  masuk_antrian.append(data_masuk_antrian)
  mulai_layanan.append(data_mulai_layanan)
  selesai_layanan.append(data_selesai_layanan)

temp = {'masuk_antrian': masuk_antrian, 'mulai_layanan': mulai_layanan, 'selesai_layanan': selesai_layanan}
data_splitted = pd.DataFrame(data=temp)
data_splitted
```

```
from datetime import datetime
import time

new_masuk_antrian = []
list_selisih_masuk_layanan_dan_selesai = []
list_selisih_masuk_antrian_dan_selesai = []
list_lama_antrian = []

for i in range(0, len(data_splitted)):
  temp = data_splitted.loc[i].values.tolist()
  
  data_masuk_antrian = datetime.strptime(temp[0].split(".")[0], '%H:%M:%S')
  data_masuk_layanan = datetime.strptime(temp[1].split(".")[0], '%H:%M:%S')
  data_selesai_layanan = datetime.strptime(temp[2].split(".")[0], '%H:%M:%S')

  selisih_masuk_antrian_dan_selesai = data_selesai_layanan - data_masuk_antrian
  selisih_layanan_dan_masuk = data_selesai_layanan - data_masuk_layanan
  selisih_lama_antrian = data_masuk_layanan - data_masuk_antrian
  
  list_selisih_masuk_layanan_dan_selesai.append(selisih_layanan_dan_masuk.total_seconds())
  list_selisih_masuk_antrian_dan_selesai.append(selisih_masuk_antrian_dan_selesai.total_seconds())
  list_lama_antrian.append(selisih_lama_antrian.total_seconds())

  if i != 0:
    temp = data_splitted.loc[i - 1].values.tolist()

    prev_data_masuk_antrian = datetime.strptime(temp[0].split(".")[0], '%H:%M:%S')

    if (data_masuk_antrian - prev_data_masuk_antrian).total_seconds() <= 5:
      prev_data_masuk_layanan = datetime.strptime(temp[1].split(".")[0], '%H:%M:%S')
      prev_data_selesai_layanan = datetime.strptime(temp[2].split(".")[0], '%H:%M:%S')

      selisih_layanan_dan_masuk = prev_data_selesai_layanan - prev_data_masuk_layanan

      jumlah_selisih_dan_masuk = selisih_layanan_dan_masuk + data_masuk_antrian
      new_masuk_antrian.append(jumlah_selisih_dan_masuk.time())
    else:
      new_masuk_antrian.append(data_masuk_antrian.time())
  else:
    new_masuk_antrian.append(data_masuk_antrian.time())

temp = {'masuk_antrian_baru': new_masuk_antrian, 'lama_layanan': list_selisih_masuk_layanan_dan_selesai, 'lama_antrian': list_lama_antrian, 'lama_proses': list_selisih_masuk_antrian_dan_selesai}
new_attributes = pd.DataFrame(data=temp)
new_attributes
```

```
data_splitted['new_masuk_antrian'] = new_attributes['masuk_antrian_baru']
data_splitted['lama_antrian'] = new_attributes['lama_antrian']
data_splitted['lama_layanan'] = new_attributes['lama_layanan']
data_splitted['lama_proses'] = new_attributes['lama_proses']
data_splitted
```

```
data_splitted.to_csv(output_path)
```
